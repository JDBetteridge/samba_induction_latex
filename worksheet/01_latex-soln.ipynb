# LaTeX Document
For the purposes of training material we use the `%%pdflatex` magic to demonstrate LaTeX code snippets. This allows us to break the document up and compile each part seperately, in order to demonstrate what each section produces. This isn't perfect, references and numbering do not (currently) work across cells, and the output occasionally looks different to what might be produces in a document.

To generate the output for each cell in this document `%%pdflatex` does:
```latex
    \documentclass[convert, varwidth]{standalone}
    \usepackage{amsmath,amssymb,amsthm,amsxtra,graphicx}
    \begin{document}
        %CELLTEXT
    \end{document}
```

You should have a document that looks something like:
```latex
    \documentclass[a4paper]{article}
    \usepackage{amsmath,amssymb,amsthm,amsxtra,graphicx}
    \begin{document}
        %All of you text
    \end{document}
```

## Information: 1: Listing things

````python
%%pdflatex
\subsection*{Listing things}

Okay, lets start with a list of my favourite things:
\begin{itemize}
    \item Raindrops on \emph{roses}
    \item Whiskers on \emph{kittens}
    \item Bright copper \emph{kettles}
    \item Warm woolen \emph{mittens}
    \item Brown paper \emph{packages} tied up with string
\end{itemize}

%%pdflatex
How to make scones:
\begin{enumerate}
    \item Heat the oven to 220C/425F/Gas 7. Lightly grease a baking sheet.
    \item
    \begin{enumerate}
        \item Mix together the \textbf{flour \& salt} and rub in the \textbf{butter}.
        \item Stir in the \textbf{sugar} \& then the \textbf{milk} to get a soft dough.
    \end{enumerate}
    \item Turn on to a floured work surface and knead very lightly. Pat out to a round 2cm/(3/4)in thick. Use a 5cm/2in cutter to stamp out rounds and place on a baking sheet. Lightly knead together the rest of the dough and stamp out more scones to use it all up.
    \item Brush the tops of the scones with the beaten egg. Bake for 12-15 minutes until well risen and golden.
    \item Cool on a wire rack and serve with butter and good \textbf{jam} and maybe some clotted cream.
\end{enumerate}

Notice the \textbf{boldface} and \emph{italics}, as well as the special symbol \&!
````

## Information: 2: Writing Maths

````python
%%pdflatex
\subsection*{Writing Maths}
Rewrite all of these, hopefully familiar equations in \LaTeX

Quadratic equation $ax^2 + bx + c = 0$.

Quadratic formula
\begin{equation}
    x = \frac{-b\pm\sqrt{b^2 - 4ac}}{2a}
\end{equation}

A matrix equation
\begin{equation}
    \begin{pmatrix}
        a & \log{b}\\
        c & d
    \end{pmatrix}
    \begin{pmatrix}
        x\\
        y
    \end{pmatrix}
    =
    \begin{pmatrix}
        0\\
        0
    \end{pmatrix}
\end{equation}

can be rewriten as a system of linear equations
\begin{align}
    ax + \log{(b)}y &= 0\nonumber\\
    cx + dy &= 0
\end{align}

Riemann zeta function
\begin{equation}
    \zeta(s) = \sum_{n=1}^\infty \frac{1}{n^s}
\end{equation}

can be rewritten as
\begin{equation}
    \zeta(s) = \frac{1}{\Gamma(s)} \int_0^\infty \frac{x^{s-1}}{e^x - 1}\ dx
\end{equation}
where
\begin{equation}
    \Gamma(s) = \int_0^\infty e^{-x}x^{s-1}\ dx
\end{equation}

- Output special characters: $S$ is a set of programming languages that are commonly used in SAMba
\begin{equation}
    S = \{\text{Matlab}, \text{R}, \text{Fortran}, \text{Python}\}
\end{equation}
- Use of the underset command: 
\begin{equation}
    x^* = \underset{x}{\text{argmin}} \, \cal{L}(x)
\end{equation}
- Control the size of brackets/parentheses:
\begin{equation}
    \left(\frac{a+b}{c+d}\right)
\end{equation}
````
